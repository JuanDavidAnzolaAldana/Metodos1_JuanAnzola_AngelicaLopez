# Punto a: Resolver analíticamente:
$$
\int_0^\infty\cdots\int_0^\infty(\frac{1}{n}\sum_{i=1}^nx_i)^2\frac{e^{-x_1/\theta}}{\theta}\cdots\frac{e^{-x_n/\theta}}{\theta}dx_1\cdots dx_n-[\int_0^\infty\cdots\int_0^\infty(\frac{1}{n}\sum_{i=1}^nx_i)\frac{e^{-x_1/\theta}}{\theta}\cdots\frac{e^{-x_n/\theta}}{\theta}dx_1\cdots dx_n]^2
$$
Al principio se va a considerar solamente la primera integral:
$$
\int_0^\infty\cdots\int_0^\infty(\frac{1}{n}\sum_{i=1}^nx_i)^2\frac{e^{-x_1/\theta}}{\theta}\cdots\frac{e^{-x_n/\theta}}{\theta}dx_1\cdots dx_n
$$
Lo primero que se puede hacer es sacar el 1/n del paréntesis:
$$
\int_0^\infty\cdots\int_0^\infty\frac{1}{n^2}(\sum_{i=1}^nx_i)^2\frac{e^{-x_1/\theta}}{\theta}\cdots\frac{e^{-x_n/\theta}}{\theta}dx_1\cdots dx_n
$$
Ahora se puede sacar el último término de la sumatoria:
$$
\int_0^\infty\cdots\int_0^\infty\frac{1}{n^2}(\sum_{i=1}^{n-1}x_i+x_n)^2\frac{e^{-x_1/\theta}}{\theta}\cdots\frac{e^{-x_n/\theta}}{\theta}dx_1\cdots dx_n
$$
Nótese que si se va a integrar con respecto a xn, la integral toma la forma de:
$$
\int_0^\infty k(x_n+s)^2\frac{e^{-x_n/\theta}}{\theta}dx_n
$$
Donde:
$$
k=\frac{1}{n^2}\frac{e^{-x_1/\theta}}{\theta}\cdots\frac{e^{-x_{n-1}/\theta}}{\theta}\\
s=\sum_{i=1}^{n-1}x_i
$$
Esta integral se puede resolver por partes:
$$
\int_0^\infty k(x_n+s)^2\frac{e^{-x_n/\theta}}{\theta}dx_n\\
\int udv=uv-\int vdu\\
u=k(x_n+s)^2\\
du=2k(x_n+s)dx_n\\
dv=\frac{e^{-x_n/\theta}}{\theta}dx_n\\
v=-e^{-x_n/\theta}\\
\int k(x_n+s)^2\frac{e^{-x_n/\theta}}{\theta}dx_n=-k(x_n+s)^2e^{-x_n/\theta}+\int 2k(x_n+s)e^{-x_n/\theta}dx_n
$$
Ahora se puede integrar por partes otra vez:
$$
\int 2k(x_n+s)e^{-x_n/\theta}dx_n\\
\int udv=uv-\int vdu\\
u=2k(x_n+s)\\
du=2kdx_n\\
dv=e^{-x_n/\theta}dx_n\\
v=-\theta e^{-x_n/\theta}\\
\int 2k(x_n+s)e^{-x_n/\theta}dx_n=-2k\theta(x_n+s)e^{-x_n/\theta}+\int2k\theta e^{-x_n/\theta}dx_n
$$
La última integral se resuelve directamente:
$$
\int2k\theta e^{-x_n/\theta}dx_n=-2k\theta^2e^{-x_n/\theta}
$$
Etonces, al reemplazar las integrales queda:
$$
\int k(x_n+s)^2\frac{e^{-x_n/\theta}}{\theta}dx_n=-k(x_n+s)^2e^{-x_n/\theta}-2k\theta(x_n+s)e^{-x_n/\theta}-2k\theta^2e^{-x_n/\theta}\\
=-ke^{-x_n/\theta}[(x_n+s)^2+2\theta(x_n+s)+2\theta^2]
$$
Para evaluar la integral en los límites, se usa la regla de l'hospital:
$$
\lim_{x_n\to\infty}-ke^{-x_n/\theta}[(x_n+s)^2+2\theta(x_n+s)+2\theta^2]\\
=\lim_{x_n\to\infty}-k\frac{(x_n+s)^2+2\theta(x_n+s)+2\theta^2}{e^{x_n/\theta}}\\
=\lim_{x_n\to\infty}-k\frac{(2(x_n+s)+2\theta)\theta}{e^{x_n/\theta}}\\
=\lim_{x_n\to\infty}-k\frac{2\theta^2}{e^{x_n/\theta}}\\
=-2k\theta^2\lim_{x_n\to\infty}e^{-x_n/\theta}\\
=-2k\theta^2(0)\\
=0
$$
Entonces, la integral evaluada es:
$$
\int_0^\infty k(x_n+s)^2\frac{e^{-x_n/\theta}}{\theta}dx_n=0+ke^{-0/\theta}[(0+s)^2+2\theta(0+s)+2\theta^2]\\
=ke^0[s^2+2s\theta+2\theta^2]=k(s+\theta)^2+k\theta^2
$$
Por lo que la integral original queda como:
$$
\int_0^\infty\cdots\int_0^\infty\frac{1}{n^2}(\sum_{i=1}^{n-1}x_i+x_n)^2\frac{e^{-x_1/\theta}}{\theta}\cdots\frac{e^{-x_n/\theta}}{\theta}dx_1\cdots dx_n=\\
\int_0^\infty\cdots\int_0^\infty\frac{1}{n^2}(\sum_{i=1}^{n-1}x_i+\theta)^2\frac{e^{-x_1/\theta}}{\theta}\cdots\frac{e^{-x_{n-1}/\theta}}{\theta}dx_1\cdots dx_{n-1}+\frac{\theta^2}{n^2}\int_0^\infty\cdots\int_0^\infty\frac{e^{-x_1/\theta}}{\theta}\cdots\frac{e^{-x_{n-1}/\theta}}{\theta}dx_1\cdots dx_{n-1}
$$
La integral de la derecha es fácil de calcular. Si se va a integrar con respecto a x(n-1), la integral toma la forma de:
$$
\int_0^\infty\cdots\int_0^\infty\frac{e^{-x_1/\theta}}{\theta}\cdots\frac{e^{-x_{n-1}/\theta}}{\theta}dx_1\cdots dx_{n-1}\\
\int k\frac{e^{-x_{n-1}/\theta}}{\theta}dx_{n-1}
$$
Donde:
$$
k=\frac{e^{-x_1/\theta}}{\theta}\cdots\frac{e^{-x_{n-2}/\theta}}{\theta}
$$
Esta integral se puede resolver directamente:
$$
\int k\frac{e^{-x_{n-1}/\theta}}{\theta}dx_{n-1}=-ke^{-x_{n-1}\theta}
$$
Entonces, la integral evaluada es:
$$
\int_0^\infty k\frac{e^{-x_{n-1}/\theta}}{\theta}dx_{n-1}=0-(-ke^0)=k\\
\int_0^\infty\cdots\int_0^\infty\frac{e^{-x_1/\theta}}{\theta}\cdots\frac{e^{-x_{n-1}/\theta}}{\theta}dx_1\cdots dx_{n-1}=\int_0^\infty\cdots\int_0^\infty\frac{e^{-x_1/\theta}}{\theta}\cdots\frac{e^{-x_{n-2}/\theta}}{\theta}dx_1\cdots dx_{n-2}
$$
Como la integral resultante tiene la misma forma de la anterior, el resultado de esta va a tener la misma forma. Si se integrara n-2 veces, quedaría lo siguiente:
$$
\int_0^\infty\cdots\int_0^\infty\frac{e^{-x_1/\theta}}{\theta}\cdots\frac{e^{-x_{n-1}/\theta}}{\theta}dx_1\cdots dx_{n-1}=\int_0^\infty\frac{e^{-x_1/\theta}}{\theta}dx_1=1
$$
Por lo tanto, la integral original se resuelve como:
$$
\int_0^\infty\cdots\int_0^\infty\frac{1}{n^2}(\sum_{i=1}^{n-1}x_i+x_n)^2\frac{e^{-x_1/\theta}}{\theta}\cdots\frac{e^{-x_n/\theta}}{\theta}dx_1\cdots dx_n=\\
\int_0^\infty\cdots\int_0^\infty\frac{1}{n^2}(\sum_{i=1}^{n-1}x_i+\theta)^2\frac{e^{-x_1/\theta}}{\theta}\cdots\frac{e^{-x_{n-1}/\theta}}{\theta}dx_1\cdots dx_{n-1}+\frac{\theta^2}{n^2}
$$
Como la integral resultante tiene la misma forma de la anterior, el resultado de esta va a tener la misma forma. Si se integrara n-1 veces, quedaría lo siguiente:
$$
\int_0^\infty\cdots\int_0^\infty\frac{1}{n^2}(\sum_{i=1}^{n-1}x_i+x_n)^2\frac{e^{-x_1/\theta}}{\theta}\cdots\frac{e^{-x_n/\theta}}{\theta}dx_1\cdots dx_n=\int_0^\infty\frac{1}{n^2}(x_1+\sum_{i=1}^{n-1}\theta)^2\frac{e^{-x_1/\theta}}{\theta}dx_1+\sum_{i=1}^{n-1}\frac{\theta^2}{n^2}\\=\frac{1}{n^2}(\sum_{i=1}^{n}\theta)^2+\sum_{i=1}^n\frac{\theta^2}{n^2}=\frac{(n\theta)^2}{n^2}+\frac{n\theta^2}{n^2}=\theta^2+\frac{\theta^2}{n}
$$
Para la segunda integral se puede hacer un proceso similar:
$$
\int_0^\infty\cdots\int_0^\infty(\frac{1}{n}\sum_{i=1}^nx_i)\frac{e^{-x_1/\theta}}{\theta}\cdots\frac{e^{-x_n/\theta}}{\theta}dx_1\cdots dx_n
$$
Se puede sacar el 1/n del paréntesis y separar el último término de la sumatoria:
$$
\int_0^\infty\cdots\int_0^\infty\frac{1}{n}(\sum_{i=1}^{n-1}x_i+x_n)\frac{e^{-x_1/\theta}}{\theta}\cdots\frac{e^{-x_n/\theta}}{\theta}dx_1\cdots dx_n
$$
Nótese que si se va a integrar con respecto a xn, la integral toma la forma de:
$$
\int k(x_n+s)\frac{e^{-x_n/\theta}}{\theta}dx_n
$$
Donde:
$$
k=\frac{1}{n}\frac{e^{-x_1/\theta}}{\theta}\cdots\frac{e^{-x_{n-1}/\theta}}{\theta}\\
s=\sum_{i=1}^{n-1}x_i
$$
Esta integral se puede resolver por partes:
$$
\int k(x_n+s)\frac{e^{-x_n/\theta}}{\theta}dx_n\\
\int udv=uv-\int vdu\\
u=k(x_n+s)\\
du=kdx_n\\
dv=\frac{e^{-x_n/\theta}}{\theta}dx_n\\
v=-e^{-x_n/\theta}\\
\int k(x_n+s)\frac{e^{-x_n/\theta}}{\theta}dx_n=-k(x_n+s)e^{-x_n/\theta}+\int ke^{-x_n/\theta}dx_n
$$
La siguiente integral se puede resolver directamente:
$$
\int ke^{-x_n/\theta}dx_n=-k\theta e^{-x_n/\theta}
$$
Etonces, al reemplazar las integrales queda:
$$
\int k(x_n+s)\frac{e^{-x_n/\theta}}{\theta}dx_n=-k(x_n+s)e^{-x_n/\theta}-k\theta e^{-x_n/\theta}=-ke^{-x_n/\theta}(x_n+s+\theta)
$$
Para evaluar la integral en los límites, se usa la regla de l'hospital:
$$
\lim_{x_n\to\infty}-ke^{-x_n/\theta}(x_n+s+\theta)\\
=\lim_{x_n\to\infty}-k\frac{(x_n+s+\theta)}{e^{x_n/\theta}}\\
=\lim_{x_n\to\infty}-k\frac{\theta(1)}{e^{x_n/\theta}}\\
=-k\theta\lim_{x_n\to\infty}e^{-x_n/\theta}\\
=-k\theta(0)\\
=0
$$
Entonces, la integral evaluada es:
$$
\int_0^\infty k(x_n+s)\frac{e^{-x_n/\theta}}{\theta}dx_n=0+ke^{-0/\theta}(0+s+\theta)=ke^0(s+\theta)=k(s+\theta)
$$
Por lo que la integral original queda como:
$$
\int_0^\infty\cdots\int_0^\infty\frac{1}{n}(\sum_{i=1}^{n-1}x_i+x_n)\frac{e^{-x_1/\theta}}{\theta}\cdots\frac{e^{-x_n/\theta}}{\theta}dx_1\cdots dx_n=\int_0^\infty\cdots\int_0^\infty\frac{1}{n}(\sum_{i=1}^{n-1}x_i+\theta)\frac{e^{-x_1/\theta}}{\theta}\cdots\frac{e^{-x_{n-1}/\theta}}{\theta}dx_1\cdots dx_{n-1}
$$
Como la integral resultante tiene la misma forma de la anterior, el resultado de esta va a tener la misma forma. Si se integrara n-1 veces, quedaría lo siguiente:
$$
\int_0^\infty\cdots\int_0^\infty\frac{1}{n}(\sum_{i=1}^{n-1}x_i+x_n)\frac{e^{-x_1/\theta}}{\theta}\cdots\frac{e^{-x_n/\theta}}{\theta}dx_1\cdots dx_n=\int_0^\infty\frac{1}{n}(x_i+\sum_{i=1}^{n-1}\theta)\frac{e^{-x_1/\theta}}{\theta}dx_1=\frac{\sum_{i=1}^{n}\theta}{n}=\frac{n\theta}{n}=\theta
$$
Entoncs, la solución analítica a la expresión es:
$$
\int_0^\infty\cdots\int_0^\infty(\frac{1}{n}\sum_{i=1}^nx_i)^2\frac{e^{-x_1/\theta}}{\theta}\cdots\frac{e^{-x_n/\theta}}{\theta}dx_1\cdots dx_n-[\int_0^\infty\cdots\int_0^\infty(\frac{1}{n}\sum_{i=1}^nx_i)\frac{e^{-x_1/\theta}}{\theta}\cdots\frac{e^{-x_n/\theta}}{\theta}dx_1\cdots dx_n]^2=(\theta^2+\frac{\theta^2}{n})-(\theta)^2\\
=\frac{\theta^2}{n}

# Punto b: Integración de montecarlo.
Se dice que si f es una función de densidad de probabilidad y g es una función se cumple que:
$$
\int_{-\infty}^\infty f(x)g(x)dx=\sum_{i=1}^N g(x_i)
$$
Entonces se puede asignar:
$$
g=(\frac{1}{n}\sum_{i=1}^nx_i)^2\\
f= \left\{ \begin{array}{lcc}
             \frac{e^{-x_1/\theta}}{\theta} & x>0 \\
             \\ 0 & x<0 \\
             \end{array}
   \right.
$$
Por lo que:
$$
\int_0^\infty(\frac{1}{n}\sum_{i=1}^nx_i)^2\frac{e^{-x_1/\theta}}{\theta}dx_1=\sum_{i=1}^N \frac{1}{n^2}(x_i+\sum_{j=2}^nx_j)^2
$$
Donde xi se escoge con la distribución aleatoria f y la sumatoria se toma como una constante.
Al hacer la integral n veces queda lo siguiente:
$$
\int_0^\infty\cdots\int_0^\infty(\frac{1}{n}\sum_{i=1}^nx_i)^2\frac{e^{-x_1/\theta}}{\theta}\cdots\frac{e^{-x_n/\theta}}{\theta}dx_1\cdots dx_n=\frac{1}{N}\sum_{i=1}^N \frac{1}{n^2}(\sum_{j=1}^nx_{ij})^2
$$
Donde xij se escoge con la distribución aleatoria f y la sumatoria se toma como una constante.
Para la segunda integral, el resultado es análogo:
$$
\int_0^\infty\cdots\int_0^\infty(\frac{1}{n}\sum_{i=1}^nx_i)\frac{e^{-x_1/\theta}}{\theta}\cdots\frac{e^{-x_n/\theta}}{\theta}dx_1\cdots dx_n=\frac{1}{N}\sum_{i=1}^N \frac{1}{n}(\sum_{j=1}^nx_{ij})^2
$$
Se puede encontrar un generador de números aleatorios que cumpla la distribución f usando la fórmula:
$$
F(x)=\int_0^x\frac{e^{-t/\theta}}{\theta}dt\\
F(x)=1-e^{-x/\theta}\\
1-F(x)=e^{-x/\theta}\\
ln(1-F(x))=-x/\theta\\
-\theta ln(1-F(x))=x\\
F^-1(x)=-\theta ln(1-x)\\
f=F^-1(uniforme)\\
f=-\theta ln(1-uniforme)\\
$$
Se va a definir y verificar la función de distribución aleatoria.

```python
import numpy
import matplotlib.pyplot as plt
def random(t,size):
    return-t*numpy.log(1-numpy.random.uniform(low=0,high=1,size=size))
def grafica(t):
    array=random(t,20000)
    x0=numpy.linspace(0,numpy.max(array),100)
    plt.plot(x0,numpy.exp(-x0/t)/t,color="r")
    plt.hist(array,density=True,bins=20)
    plt.show()
grafica(0.5)
grafica(1)
grafica(2)
```

Como la función de distribución aleatoria funciona, se puede proceder a calcular la primera integral:

```python
import tqdm
theta=2
dim=20
n=10**6
integral1=0
integral2=0
for i in tqdm.trange(n):
    x=random(theta,dim)
    promedio=numpy.sum(x)/dim
    integral1+=promedio**2
    integral2+=promedio
integral1/=n
integral2/=n
print("La primera integral es ",integral1)
print("La segunda integral es ",integral2)
print("La varianza es ",integral1-(integral2**2))
```
