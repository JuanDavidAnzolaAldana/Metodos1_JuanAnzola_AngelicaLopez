```python
#Punto 2.03: Máximos locales
import urllib.request
import numpy as np
import matplotlib.pyplot as plt

url = 'https://raw.githubusercontent.com/asegura4488/Database/main/MetodosComputacionalesReforma/EstrellaEspectro.txt'
filename = 'EstrellaEspectro.txt'

urllib.request.urlretrieve(url, filename)
data = np.loadtxt(filename)
plt.plot(data[:,0],data[:,1])
```

```python
#Comparar puntos
x = np.array(data[:,0])
y = np.array(data[:,1])
n = len(y)

ymax = []
xmax=[]

for j in range(1, n-1):
    if y[j-1] < y[j] and y[j] > y[j+1]:
        ymax.append(y[j])
        xmax.append(x[j])

plt.plot(data[:,0],data[:,1])
plt.scatter(xmax,ymax, color='red')
```
