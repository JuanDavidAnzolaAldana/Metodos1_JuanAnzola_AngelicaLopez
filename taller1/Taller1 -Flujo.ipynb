```python
#Punto 3.6
%matplotlib notebook
import numpy as np
import matplotlib.pyplot as plt
#Calcular el campo de velocidades cerca de la superficie de un cilindro de radio R = 2cm 
#a) Discretización donde la región es: A ∈ [−4, 4] con 25 puntos en cada eje.

x = np.linspace(-4,4,25)
y = np.linspace(-4,4,25)
X,Y = np.meshgrid(x,y)
R = 2 #cm
v = 2 #cm/s

#b) Función potencial del flujo

def Getpf(x,y):
    pf = v*x*(1 - (R**2/(x**2 + y**2)))
    return pf

#c) Derivadas parciales

def Getvx(x,y):

    h = 0.001
    dfdx = 0.
  
    dfdx = (Getpf(x+h,y) - (Getpf(x-h,y)))/2*h

    return dfdx

def Getvy(x,y):

    h = 0.001
    dfdy = 0.

    dfdy = (Getpf(x,y+h) - (Getpf(x,y-h)))/2*h

    return dfdy

```

```python
#Graficar
n = len(x)

Vx = np.zeros((n,n))
Vy = np.zeros((n,n))

for i in range(0,n):
    for j in range(0,n):
        Vx[i,j] = Getvx(x[i],y[j])
        Vy[i,j] = Getvy(x[i],y[j])


for i in range(0,n):
    for j in range(0,n):
        if (x[i]**2 + y[j]**2) > R**2:
            plt.quiver(x[i],y[j],Vx[i,j],Vy[i,j])
plt.show()
```
